```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data
df=pd.read_csv("H:/sanjeev/Diwali Sales Data.csv",encoding="unicode_escape")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Removing unnecessary columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#checking null values
df.isnull().sum()
```

```python
#filling null values by zero
df["Amount"].fillna(value=0,inplace=True)
df.isnull().sum()
```

```python
#changing type of data of amount columns
df["Amount"]=df["Amount"].astype("int")
df["Amount"].dtype
```

```python
df.columns
```

```python
df.rename(columns={"Marital_Status":"Shadi"},inplace=True)
df
```

```python
df.describe()
```

# exploratory analysis

```python
#chart one
c=sns.countplot(x="Gender",data=df)
plt.title("count plot of gender")
for bars in c.containers:
    c.bar_label(bars)
```

```python
#chart two
sales_by_gender=df.groupby(["Gender"],as_index=False) ["Amount"].sum()
sns.barplot(x="Gender",y="Amount",data=sales_by_gender)
plt.title("Sales by gender")
plt.show()
```

```python
#chart three
a=sns.countplot(x="Age Group",hue="Gender",data=df)
for bars in a.containers:
    a.bar_label(bars)
plt.title("Age group across gender")
```

```python
#chart four
j=df.groupby(["Age Group"], as_index=False) ["Amount"].sum().sort_values(by="Amount",ascending=False)
br=sns.barplot(x="Age Group",y="Amount",data=j)
plt.title("Total sales by age")
plt.show()
```

```python
df.columns
```

```python
#chart five
s=df.groupby(["State"],as_index=False) ["Orders"].sum().sort_values(by="Orders",ascending=False)
plt.figure(figsize=(22,8))
g=sns.barplot(x="State",y="Orders",data=s)
plt.title("Orders by state")
for bars in g.containers:
    g.bar_label(bars)
plt.show()
```

```python
#chart six
am=df.groupby(["State"],as_index=False) ["Amount"].sum().sort_values(by="Amount",ascending=False)
plt.figure(figsize=(20,5))
sns.barplot(x="State",y="Amount",data=am)
```

```python
#Chart seven
plt.figure(figsize=(6,6))
a=sns.countplot(x="Gender",data=df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
#Chart eight
d=df.groupby(["Shadi","Gender"], as_index=False) ["Amount"].sum().sort_values(by="Amount", ascending=False)
a=sns.barplot(data=d,x="Shadi",y="Amount",hue="Gender")
plt.title("Sales by marital status")
```

```python
#Chart nine
a=df.groupby(["Occupation"],as_index=False) ["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.figure(figsize=(15,6))
sns.barplot(x="Occupation",y="Amount",data=a)
plt.title("Top 10 occupation Sales")
```
